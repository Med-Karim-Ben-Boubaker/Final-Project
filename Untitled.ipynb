As we can see here, this data can be exploited by transforming object columns into numerical values.
First, we are going to add unities to the Column names to describe the values after the transformation

```python
# Adding unities to some column for clarification in the data vazialaztion process
df.rename(columns = {'Acceleration': 'Acceleration(sec)', 'TopSpeed': 'TopSpeed(km/h)',
                     'Range': 'Range(km)', 'Efficiency': 'Efficiency(Wh/km)', 'FastChargeSpeed': 'FastChargeSpeed(km/h)',
                     'PriceinGermany': 'PriceinGermany(€)', 'PriceinUK': 'PriceinUK(£)'}, inplace = True)
df.columns
```

changing values of this DataFrame  from string values into numerical one 

```python
import re  # Importing re function for extracting values 

def extract_values(column):  # Transforming values into numerical data 
    for i in range(180):
        try:
            if ('€' in df[column][i]) or ('£' in df[column][i]): # Transforming prices numerical value 
                var = re.findall(r'\d+', df[column][i])          # ex: €40,000 --> 40000
                df[column][i] = int(var[0])*1000
            else:   
                var = re.findall(r'\d+', df[column][i])          # Transforming other data 
                df[column][i] = int(var[0])
            
        except:
            df[column][i] = np.nan   # If element contains no value then value = NaN
            


li = ['TopSpeed(km/h)', 'Range(km)', 'FastChargeSpeed(km/h)', 'Efficiency(Wh/km)', 'PriceinGermany(€)', 'PriceinUK(£)']
for column in li:
    extract_values(column) # Applying the function
```

We can exploit the name column by extracting the power and turn it into numerical value in a new column called (power(kWh)).

```python
po = []
for i in range(len(df['Subtitle'])): # Adding a column named power based on Subtitle
    var = int(re.findall(r'\d+', df['Subtitle'][i])[0])
    po.append(var)
df['Power(kWh)'] = po
    
for i in range(len(df['Acceleration(sec)'])): # Exception case (values are float type)
    var = float(re.findall("\d+\.\d+", df['Acceleration(sec)'][i])[0])
    df['Acceleration(sec)'][i] = var
```

Extracting Manufacturer from name Column

```python
man = [] # a list to store the Manufacturer names to put it later into a new column
for i in range(len(df['Name'])):
    var = df['Name'][i].split()[0]
    man.append(var)

df['Manufacturer'] = man
```

If we drop all the missing rows of PriceinGermany and PriceinUK, it'll be too much rows to drop. So, we can fill the missing values in PriceinGermany with PriceinUK 0.86 and PriceinUK with PriceinGermany 1.17.

*at the time i'm doing eda on this dataset, euro to pound sterling was 0.86 and pound sterling to euro was 1.17

```python
df['PriceinGermany(€)'].fillna(df['PriceinUK(£)'] * 0.85, axis=0, inplace=True)
df['PriceinUK(£)'].fillna(df['PriceinGermany(€)'] * 1.17, axis=0, inplace=True)
```

now let's fill the Nan values with the mean() function

```python
df['PriceinGermany(€)'].fillna(df['PriceinUK(£)'].mean(), inplace=True)
df['PriceinUK(£)'].fillna(df['PriceinGermany(€)'].mean(), inplace=True)
```

```python
df=df.apply(pd.to_numeric, errors='ignore')
```

```python
li = ['TopSpeed(km/h)','FastChargeSpeed(km/h)']
for column in li:
    df[column].fillna(df[column].mean(), inplace=True)
```

## Linear Rgression

```python
#extract x and y from our data
x=df['TopSpeed(km/h)'].values # We will use this column since it has the most correlation with the meanPrice
y=df['PriceinGermany(€)'].values
li=[]

x_train,x_test,y_train,y_test=train_test_split(x.reshape(-1, 1),y.reshape(-1, 1),test_size=0.35,random_state=189) #splitting data with test size of 35%
model=LinearRegression() #build linear regression model
model.fit(x_train,y_train.reshape(-1, 1))  #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance


print("R squared", metrics.r2_score(y_test,predicted))
```

```python
plt.figure(figsize = (12,7))
plt.scatter(x.reshape(-1, 1),y.reshape(-1, 1),color="r")
plt.title("Linear Regression")
plt.ylabel("PriceinGermany(€)")
plt.xlabel("TopSpeed(km/h)")
plt.plot(x.reshape(-1, 1),model.predict(x.reshape(-1, 1)),color="k")
plt.show()
```

## Ranking cars

To rank these electric cars according to the standard of price and quality, we must agree on the important criteria first, by adding a specific coefficient to each feature.

we will assume that efficiency is the most important feature and Range is the least one.

```python
a = 5 
# for efficiency
b = 4
# for FastChargeSpeed
c = 3
# for TopSpeed
d = 2
# for Acceleration
e = 1
# for Range
```

```python
df.columns
```

```python
# Calculating the Score 
score = df['Efficiency(Wh/km)']*a +df['FastChargeSpeed(km/h)']*b + df['TopSpeed(km/h)']*c + df['Acceleration(sec)']*d + df['Range(km)']*e


df["Rating"] = score
df["Rating"].head()
```

```python
rating_w_price = df["Rating"]**2/df["PriceinGermany(€)"] # Calculating the Rating with Price 


df["Price_Rating"] = rating_w_price
df["Price_Rating"].head()
```

## Sorting the best scored cars

```python
# Sorting The Best Scored Cars
scores = df["Price_Rating"]
best_scored_cars = []
best_scores = sorted(list(scores), reverse=True)[:10]
for i in range(len(best_scores)):
    best_scored_cars.append(df.loc[df['Price_Rating'] == best_scores[i]]['Name'])  
```

```python
for i in range(len(best_scored_cars)):
    
    best_scored_cars[i] = str(best_scored_cars[i]).replace('\nName: Name, dtype: object','').split()
    best_scored_cars[i]=best_scored_cars[i][1:len(best_scored_cars[i])]
    var=''
    for j in range (len(best_scored_cars[i])):
        var = var + best_scored_cars[i][j] + ' ' 
    best_scored_cars[i] = var

    best_scored_cars[i] 
```

```python
plt.figure(figsize = (12,7))
ax = sns.barplot(x=best_scores, y=best_scored_cars, palette="Blues_d")
```

### Clustring with K-means

```python
plt.scatter(df2['Age'],df2['Annual Income(€)'])
plt.show()
```

```python
x = df2[['Annual Income(€)', 'Genre(Numerical)']]
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.figure(figsize = (12,7))
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```
